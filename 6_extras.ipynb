# Extra utilities

### Extract all the Sentinel-1 dates for a given AOI

```python

import ee
ee.Initialize(project="ee-dfgm2006")
```

```python
aoi = ee.Geometry.Polygon(
        [[[108.34057909514411, -7.022847131010858],
          [108.34057909514411, -7.1809266528305615],
          [108.53146654631598, -7.1809266528305615],
          [108.53146654631598, -7.022847131010858]]])
```

```python
orbits = ["ASCENDING", "DESCENDING"]
ic = (
        ee.ImageCollection("COPERNICUS/S1_GRD_FLOAT")
        .filterBounds(aoi)
        .filterMetadata("resolution_meters", "equals", 10)
        .filter(ee.Filter.eq("instrumentMode", "IW"))
        .filter(
            ee.Filter.And(
                ee.Filter.listContains("transmitterReceiverPolarisation", "VV"),
                ee.Filter.listContains("transmitterReceiverPolarisation", "VH"),
            )
        )
        .filter(
            ee.Filter.Or(
                [ee.Filter.eq("orbitProperties_pass", orbit) for orbit in orbits]
            )
        )
    )
```

```python
def get_date(img):
    return ee.Feature(
        None, 
        ee.Dictionary(
            {
                "date" : ee.Date(img.get("system:time_start")).format("YYYY-MM-dd"),
                "orbit" : img.get("orbitProperties_pass")
            }
        )
    )

all_dates = ic.map(get_date).toList(ic.size()).map(
    lambda f: [ee.Feature(f).get("date"), ee.Feature(f).get("orbit")]
).getInfo()
```
